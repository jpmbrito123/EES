# Pyperformance - Análise de Resultados


```python
import seaborn as sea
import scipy.stats as sstats
import matplotlib.pyplot as plt
import pandas as pd

file_name = 'csv/measurementscompararpython.csv'
names = ['Language', 'Program', 'PowerLimit', 'Package', 'Core(s)', 'GPU', 'DRAM', 'Time (ms)', 'Temperature', 'Memory']
pf = pd.read_csv(file_name, header = 0, delimiter=",", decimal=".", names=names)

shape = pf.shape
columns = pf.columns

print("#> (rows:columns) => (", shape[0], ":", shape[1], ')', sep='')
print("\n#> Got the following columns:")
for col in columns:
    print("\t\t\t•", col)
```

# Tratamento de Dados
Devido à presença de valores negativos na coluna package e Core(s), o grupo decidiu remover as linhas onde esses valores se encontravam negativos.



```python
df = pf[pf['Package'] >= 0]
df = pf[pf['Core(s)'] >= 0]
# Exibir os valores únicos na coluna 'Package'

```

## Distribuição das Versões

A figura abaixo exibe a distribuição das versões do Python no conjunto de dados. Utilizando uma plotagem de barras, a contagem de cada programa é representada no eixo horizontal (x-axis), permitindo uma visualização rápida e intuitiva das frequências relativas de cada programa. 


```python
plt.figure(figsize = (10,8))
sea.countplot(data = df, x = 'Language')
```

## Distribuição de Versões por Temperatura

A figura abaixo apresenta a distribuição das versões com base na temperatura.

 Utilizando uma plotagem de barras com coloração diferenciada (hue) para as diferentes faixas de temperatura, esta visualização permite observar como a distribuição das versões varia em relação às condições de temperatura. A paleta de cores "rainbow" foi utilizada para destacar visualmente as diferentes categorias de temperatura.


```python
sea.countplot(x='Language', hue='Temperature', data=df, palette='rainbow')
```

## Relação entre Tempo de Execução, Cores e Versões

A figura abaixo exibe um gráfico de dispersão que representa a relação entre o tempo de execução (em milissegundos) e o número de cores utilizados por versão do Python. 

Este tipo de visualização permite identificar padrões ou tendências na relação entre o tempo de execução, o número de núcleos e os programas específicos.

```python
plt.figure(figsize=(10,6))
sea.scatterplot(data=df, x='Time (ms)', y='Core(s)', hue='Language')
```

## Relações Multivariadas entre Variáveis por Versão

A figura abaixo apresenta um pairplot que exibe as relações multivariadas entre as variáveis do conjunto de dados, com cada ponto colorido de acordo com a versão do python correspondente. 

Este tipo de visualização permite explorar as relações entre todas as combinações de variáveis, enquanto também identifica padrões distintos para cada programa. Observando as distribuições e dispersões dos dados, é possível ganhar insights sobre as relações complexas entre as variáveis para diferentes programas.


```python
sea.pairplot(data=df, hue='Language')
plt.show()
```

## Análise do Comportamento de Tempo de Execução e Consumo de Energia 

O gráfico abaixo apresenta uma análise do comportamento de tempo de execução e consumo de energia dos programas em diferentes versões do Python. 

Cada ponto no gráfico representa uma execução do programa, onde o eixo x representa o tempo de execução em milissegundos e o eixo y representa o consumo de energia em pacotes. Esta visualização permite uma rápida compreensão da relação entre o tempo de execução e o consumo de energia para o programa.


```python

import matplotlib.pyplot as plt

df_cleaned = df[df['Package'] >= 0]

sea.scatterplot(data=df_cleaned, x='Package', y='Time (ms)', hue='Language')
plt.title(f'Comportamento (runtime+energia) para os programas em diferentes versões do Python')
```

## Relação entre PowerLimit e Tempo de Execução por Versão


A figura abaixo apresenta um gráfico de dispersão que mostra a relação entre o limite de potência (PowerLimit) e o tempo de execução (em milissegundos) para os programas em diferentes versões do Python. 

Cada ponto no gráfico representa uma observação do conjunto de dados, colorida de acordo com a versão correspondente. 

Esta visualização possibilita a identificação de padrões ou tendências na relação entre o limite de potência e o tempo de execução para cada versão individualmente.

```python

sea.scatterplot(data=df, x='PowerLimit', y='Time (ms)', hue='Language')
plt.title(f'Comportamento (runtime+PowerLimit) para as diferentes versões do Python')
```

```python
# Calculando os menores tempos para cada PowerLimit para cada programa
min_cons = df.groupby(['Language','Program', 'PowerLimit'])['Time (ms)'].min().reset_index()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Exibindo os menores tempos
print(min_cons)
```

## Relação entre PowerLimit e Consumo de energia por Versão


A figura abaixo apresenta um gráfico de dispersão que mostra a relação entre o limite de potência (PowerLimit) e o consumo de energia (em pacotes) para os programas em diferentes versões do Python. 

Cada ponto no gráfico representa uma observação do conjunto de dados, colorida de acordo com a versão correspondente. 

Esta visualização possibilita a identificação de padrões ou tendências na relação entre o limite de potência e o consumo de energia para cada versão individualmente.

```python
# Remover as linhas onde a coluna 'Package' tem valores negativos
df_cleaned = df[df['Package'] >= 0]

# Criar o gráfico scatterplot com os dados limpos
plt.figure(figsize=(10, 6))
sea.scatterplot(data=df_cleaned, x='PowerLimit', y='Package', hue='Language')

# Adicionar título ao gráfico
plt.title(f'Comportamento (Package+PowerLimit) para as diferentes versões do Python')

# Mostrar o gráfico
plt.show()

```

```python
# Calculando os menores consumos para cada PowerLimit para cada programa
min_cons = df.groupby(['Language','Program', 'PowerLimit'])['Package'].min().reset_index()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Exibindo os menores tempos
print(min_cons)
```

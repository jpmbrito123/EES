# Dacapo - Análise de Resultados


```python
import seaborn as sea
import scipy.stats as sstats
import matplotlib.pyplot as plt
import pandas as pd

file_name = 'csv/measurementsdacapo.csv'
names = ['Language', 'Program', 'PowerLimit', 'Package', 'Core(s)', 'GPU', 'DRAM', 'Time (ms)', 'Temperature', 'Memory']
df = pd.read_csv(file_name, header = 0, delimiter=",", decimal=".", names=names)

shape = df.shape
columns = df.columns

print("#> (rows:columns) => (", shape[0], ":", shape[1], ')', sep='')
print("\n#> Got the following columns:")
for col in columns:
    print("\t\t\t•", col)
```

## Distribuição de Programas

A figura abaixo exibe a distribuição dos programas no conjunto de dados. Utilizando uma plotagem de barras, a contagem de cada programa é representada no eixo horizontal (x-axis), permitindo uma visualização rápida e intuitiva das frequências relativas de cada programa. 


```python
plt.figure(figsize = (10,8))
sea.countplot(data = df, x = 'Program')
```

## Distribuição de Programas por Temperatura

A figura abaixo apresenta a distribuição dos programas com base na temperatura.

 Utilizando uma plotagem de barras com coloração diferenciada (hue) para as diferentes faixas de temperatura, esta visualização permite observar como a distribuição dos programas varia em relação às condições de temperatura. A paleta de cores "rainbow" foi utilizada para destacar visualmente as diferentes categorias de temperatura.


```python
sea.countplot(x='Program', hue='Temperature', data=df, palette='rainbow')
```

## Relação entre Tempo de Execução, Cores e Programas

A figura abaixo exibe um gráfico de dispersão que representa a relação entre o tempo de execução (em milissegundos) e o número de cores utilizados por programa. 

Este tipo de visualização permite identificar padrões ou tendências na relação entre o tempo de execução, o número de núcleos e os programas específicos.

```python
plt.figure(figsize=(10,6))
sea.scatterplot(data=df, x='Time (ms)', y='Core(s)', hue='Program')
```

## Relações Multivariadas entre Variáveis por Programa

A figura abaixo apresenta um pairplot que exibe as relações multivariadas entre as variáveis do conjunto de dados, com cada ponto colorido de acordo com o programa correspondente. 

Este tipo de visualização permite explorar as relações entre todas as combinações de variáveis, enquanto também identifica padrões distintos para cada programa. Observando as distribuições e dispersões dos dados, é possível ganhar insights sobre as relações complexas entre as variáveis para diferentes programas.


```python
sea.pairplot(data=df, hue='Program')
plt.show()
```

## Análise do Comportamento de Tempo de Execução e Consumo de Energia 

O gráfico abaixo apresenta uma análise do comportamento de tempo de execução e consumo de energia para os programas presentes na benchmark "dacapo". 

Cada ponto no gráfico representa uma execução do programa, onde o eixo x representa o tempo de execução em milissegundos e o eixo y representa o consumo de energia em pacotes. Esta visualização permite uma rápida compreensão da relação entre o tempo de execução e o consumo de energia para o programa "dacapo".



```python

import matplotlib.pyplot as plt



plt.scatter(x=df['Time (ms)'], y=df['Package'])
plt.xlabel('Time (ms)')
plt.ylabel('Package')
plt.title(f'Comportamento (runtime+energia) para os programas presentes na benchmark "dacapo"')
plt.show()
```

## Relação entre PowerLimit e Tempo de Execução por Programa


A figura abaixo apresenta um gráfico de dispersão que mostra a relação entre o limite de potência (PowerLimit) e o tempo de execução (em milissegundos) para diferentes programas. 

Cada ponto no gráfico representa uma observação do conjunto de dados, colorida de acordo com o programa correspondente. 

Esta visualização possibilita a identificação de padrões ou tendências na relação entre o limite de potência e o tempo de execução para cada programa individualmente.

```python

sea.scatterplot(data=df, x='PowerLimit', y='Time (ms)', hue='Program')
plt.title(f'Comportamento (runtime+PowerLimit) para os Programas do dacapo')
```

```python
# Calculando os menores tempos para cada PowerLimit para cada programa
min_times = df.groupby(['Program', 'PowerLimit'])['Time (ms)'].min().reset_index()

# Exibindo os menores tempos
print(min_times)
```
